# Mask Selection Tool (Used After the Other Tool)


## Tool Description 
This tool is used to mark and select high-quality cell colonies from the inverse binary mask (Threshold Mask) that was generated using the Image Thresholding tool. Good cell colony masks can be selected by drawing a polygon boundary around it. The selected regions will be saved and output as a Segmentation Mask. <br> 

The tool takes only one input:
- The path to the color image.

The output of this tool will be:
- A binary segmentation mask, where all cell colonies are white and the background is black.

## Tool Features
- The tool will display 4 images: 
    1. Top Left: the 'Color Image' for reference when drawing your polygon.<br>
    2. Top Right: 'Thresholded Mask', the inverse binary mask generated using the Image Thresholding tool.
    3. Bottom Left: a 'Preview Mask' of the region that you have just selected.
    4. Bottom Right: the 'Segmentation Mask, where all the selected region will be stored. <br><br>
    
- You can draw a red polygon around a cell colony that you would like to select. Polygon annotation will only work on the 'Thresholded Mask'.
- The tool has a Refresh button to clear all polygon drawings and the preview mask.
- The tool has a Preview button to preview a selected cell colony/region, for you to double check before merging with the segmentation mask. If you are not satisfied with it, click the [Refresh] button to draw the polygon again.
- The tool has a [Save] button to add the selected cell colony/region to the segmentation mask and write the segmentation mask on your computer.


## Tool Usage
To load and use this tool, please follow the steps outlined below in this notebook file. Once the tool has been configured with the proper inputs (see Cell #3 below), you may use the tool as follows:

1. Simply right click your mouse on the Threshold Mask to draw a red polygon boundary around colonies. At least two right clicks must be made before the polygon is displayed. Remember, polygons can only be drawn in the Threshold Mask.<br>
2. Once a polygon has been drawn around the selected cell colony region, use the preview button to check if the selected cell colony masks look good.
3. If the preview looks good, you can add that selected region to the segmentation mask by clicking the save button. If however, you do not like the current preview, you can clear the selection by clicking the [Refresh] button.
4. After saving your selection, also click the [Refresh] button so that you can move on to select a new cell colony region.
